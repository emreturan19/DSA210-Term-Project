```python
# ============================================================
# BIST STOCK ANALYSIS PROJECT
# Advanced EDA + Hypothesis Testing + ML Prediction
# Stocks: THYAO, GARAN, ASELS, KCHOL, TUPRS, BIMAS
# Benchmark: BIST100
# ============================================================

!pip install yfinance ta statsmodels scipy scikit-learn seaborn plotly -q

import warnings
warnings.filterwarnings("ignore")

import yfinance as yf
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

from scipy import stats
from statsmodels.tsa.stattools import adfuller

from sklearn.model_selection import train_test_split, TimeSeriesSplit, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression

import ta

sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (14, 6)


# ============================================================
# 1. DATA COLLECTION
# ============================================================

stocks = {
    "THYAO": "THYAO.IS",
    "GARAN": "GARAN.IS",
    "ASELS": "ASELS.IS",
    "KCHOL": "KCHOL.IS",
    "TUPRS": "TUPRS.IS",
    "BIMAS": "BIMAS.IS",
    "BIST100": "XU100.IS"
}

start_date = "2020-01-01"
end_date = None

raw_data = yf.download(
    list(stocks.values()),
    start=start_date,
    end=end_date,
    auto_adjust=True,
    group_by="ticker",
    progress=False
)

print("Data downloaded successfully.")


# ============================================================
# 2. CLEAN CLOSE PRICE DATA
# ============================================================

close_prices = pd.DataFrame()

for name, ticker in stocks.items():
    try:
        close_prices[name] = raw_data[ticker]["Close"]
    except:
        print(f"Problem with {name}")

close_prices = close_prices.dropna()

returns = close_prices.pct_change().dropna()
log_returns = np.log(close_prices / close_prices.shift(1)).dropna()

stock_names = ["THYAO", "GARAN", "ASELS", "KCHOL", "TUPRS", "BIMAS"]

print(close_prices.head())
print("\nMissing values:")
print(close_prices.isnull().sum())


# ============================================================
# 3. NORMALIZED PRICE PERFORMANCE
# ============================================================

normalized_prices = close_prices / close_prices.iloc[0] * 100

fig = px.line(
    normalized_prices,
    x=normalized_prices.index,
    y=normalized_prices.columns,
    title="Normalized Price Performance: Stocks vs BIST100",
    labels={"value": "Normalized Price Index (Start = 100)", "index": "Date"}
)
fig.show()


# ============================================================
# 4. CUMULATIVE RETURNS
# ============================================================

cumulative_returns = (1 + returns).cumprod() - 1

fig = px.line(
    cumulative_returns,
    x=cumulative_returns.index,
    y=cumulative_returns.columns,
    title="Cumulative Returns of Stocks and BIST100",
    labels={"value": "Cumulative Return", "index": "Date"}
)
fig.show()


# ============================================================
# 5. RETURN DISTRIBUTIONS
# ============================================================

for stock in stock_names:
    fig = px.histogram(
        returns,
        x=stock,
        nbins=80,
        marginal="box",
        title=f"Daily Return Distribution with Boxplot: {stock}",
        labels={stock: "Daily Return"}
    )
    fig.show()


# ============================================================
# 6. RISK RETURN SUMMARY
# ============================================================

summary = pd.DataFrame(index=stock_names)

summary["Average Daily Return"] = returns[stock_names].mean()
summary["Daily Volatility"] = returns[stock_names].std()
summary["Annualized Return"] = returns[stock_names].mean() * 252
summary["Annualized Volatility"] = returns[stock_names].std() * np.sqrt(252)
summary["Sharpe Ratio"] = summary["Annualized Return"] / summary["Annualized Volatility"]
summary["Skewness"] = returns[stock_names].skew()
summary["Kurtosis"] = returns[stock_names].kurtosis()
summary["Minimum Daily Return"] = returns[stock_names].min()
summary["Maximum Daily Return"] = returns[stock_names].max()

display(summary.sort_values("Sharpe Ratio", ascending=False))


fig = px.scatter(
    summary,
    x="Annualized Volatility",
    y="Annualized Return",
    text=summary.index,
    size="Sharpe Ratio",
    title="Risk-Return Map",
    labels={
        "Annualized Volatility": "Annualized Risk / Volatility",
        "Annualized Return": "Annualized Return"
    }
)
fig.update_traces(textposition="top center")
fig.show()


# ============================================================
# 7. CORRELATION HEATMAP
# ============================================================

corr_matrix = returns[stock_names + ["BIST100"]].corr()

plt.figure(figsize=(12, 7))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Daily Returns")
plt.show()


# ============================================================
# 8. ROLLING VOLATILITY
# ============================================================

rolling_volatility = returns[stock_names].rolling(window=30).std() * np.sqrt(252)

fig = px.line(
    rolling_volatility,
    x=rolling_volatility.index,
    y=rolling_volatility.columns,
    title="30-Day Rolling Annualized Volatility",
    labels={"value": "Annualized Volatility", "index": "Date"}
)
fig.show()


# ============================================================
# 9. DRAWDOWN ANALYSIS
# ============================================================

def calculate_drawdown(price_series):
    running_max = price_series.cummax()
    drawdown = price_series / running_max - 1
    return drawdown

drawdowns = close_prices[stock_names].apply(calculate_drawdown)

fig = px.line(
    drawdowns,
    x=drawdowns.index,
    y=drawdowns.columns,
    title="Drawdown Analysis",
    labels={"value": "Drawdown", "index": "Date"}
)
fig.show()

max_drawdown = drawdowns.min().sort_values()
print("Maximum Drawdowns:")
display(max_drawdown)


# ============================================================
# 10. BETA AGAINST BIST100
# ============================================================

beta_results = {}

market_return = returns["BIST100"]

for stock in stock_names:
    covariance = np.cov(returns[stock], market_return)[0][1]
    market_variance = np.var(market_return)
    beta = covariance / market_variance
    beta_results[stock] = beta

beta_df = pd.DataFrame.from_dict(beta_results, orient="index", columns=["Beta vs BIST100"])
display(beta_df.sort_values("Beta vs BIST100", ascending=False))


# ============================================================
# 11. HYPOTHESIS TESTING
# H0: Stock average daily return = BIST100 average daily return
# H1: Stock average daily return != BIST100 average daily return
# ============================================================

hypothesis_results = []

for stock in stock_names:
    t_stat, p_value = stats.ttest_ind(
        returns[stock],
        returns["BIST100"],
        equal_var=False
    )

    result = "Reject H0" if p_value < 0.05 else "Fail to Reject H0"

    hypothesis_results.append({
        "Stock": stock,
        "Mean Return": returns[stock].mean(),
        "BIST100 Mean Return": returns["BIST100"].mean(),
        "t-statistic": t_stat,
        "p-value": p_value,
        "Decision at 5%": result
    })

hypothesis_df = pd.DataFrame(hypothesis_results)
display(hypothesis_df)


# ============================================================
# 12. NORMALITY TEST
# H0: Returns are normally distributed
# ============================================================

normality_results = []

for stock in stock_names:
    stat, p_value = stats.jarque_bera(returns[stock])

    normality_results.append({
        "Stock": stock,
        "Jarque-Bera Statistic": stat,
        "p-value": p_value,
        "Decision": "Reject normality" if p_value < 0.05 else "Fail to reject normality"
    })

normality_df = pd.DataFrame(normality_results)
display(normality_df)


# ============================================================
# 13. STATIONARITY TEST
# H0: Series has a unit root / non-stationary
# ============================================================

adf_results = []

for stock in stock_names:
    adf_stat, p_value, _, _, _, _ = adfuller(returns[stock])

    adf_results.append({
        "Stock": stock,
        "ADF Statistic": adf_stat,
        "p-value": p_value,
        "Decision": "Stationary" if p_value < 0.05 else "Non-stationary"
    })

adf_df = pd.DataFrame(adf_results)
display(adf_df)


# ============================================================
# 14. FEATURE ENGINEERING FOR MACHINE LEARNING
# Target: Predict whether THYAO next-day return is positive
# ============================================================

target_stock = "THYAO"

ml_data = pd.DataFrame(index=close_prices.index)

ml_data["Close"] = close_prices[target_stock]
ml_data["Return"] = returns[target_stock]
ml_data["Market_Return"] = returns["BIST100"]

ml_data["Lag_1"] = ml_data["Return"].shift(1)
ml_data["Lag_2"] = ml_data["Return"].shift(2)
ml_data["Lag_3"] = ml_data["Return"].shift(3)
ml_data["Lag_5"] = ml_data["Return"].shift(5)

ml_data["Rolling_Return_5"] = ml_data["Return"].rolling(5).mean()
ml_data["Rolling_Return_10"] = ml_data["Return"].rolling(10).mean()
ml_data["Rolling_Volatility_5"] = ml_data["Return"].rolling(5).std()
ml_data["Rolling_Volatility_10"] = ml_data["Return"].rolling(10).std()

ml_data["SMA_10"] = ml_data["Close"].rolling(10).mean()
ml_data["SMA_30"] = ml_data["Close"].rolling(30).mean()
ml_data["SMA_Ratio"] = ml_data["SMA_10"] / ml_data["SMA_30"]

ml_data["RSI"] = ta.momentum.RSIIndicator(ml_data["Close"], window=14).rsi()
ml_data["MACD"] = ta.trend.MACD(ml_data["Close"]).macd()
ml_data["MACD_Signal"] = ta.trend.MACD(ml_data["Close"]).macd_signal()

ml_data["Target"] = (ml_data["Return"].shift(-1) > 0).astype(int)

ml_data = ml_data.dropna()

display(ml_data.head())


# ============================================================
# 15. TRAIN TEST SPLIT
# Time-series style split: no random shuffle
# ============================================================

features = [
    "Market_Return",
    "Lag_1", "Lag_2", "Lag_3", "Lag_5",
    "Rolling_Return_5", "Rolling_Return_10",
    "Rolling_Volatility_5", "Rolling_Volatility_10",
    "SMA_Ratio",
    "RSI",
    "MACD",
    "MACD_Signal"
]

X = ml_data[features]
y = ml_data["Target"]

split_index = int(len(ml_data) * 0.8)

X_train = X.iloc[:split_index]
X_test = X.iloc[split_index:]

y_train = y.iloc[:split_index]
y_test = y.iloc[split_index:]

print("Training size:", X_train.shape)
print("Test size:", X_test.shape)


# ============================================================
# 16. ML MODELS
# ============================================================

models = {
    "Logistic Regression": Pipeline([
        ("scaler", StandardScaler()),
        ("model", LogisticRegression(max_iter=1000))
    ]),

    "Random Forest": RandomForestClassifier(
        n_estimators=300,
        max_depth=5,
        random_state=42
    ),

    "Gradient Boosting": GradientBoostingClassifier(
        n_estimators=200,
        learning_rate=0.05,
        max_depth=3,
        random_state=42
    )
}

model_results = []

for name, model in models.items():
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)

    if hasattr(model, "predict_proba"):
        probabilities = model.predict_proba(X_test)[:, 1]
    else:
        probabilities = predictions

    accuracy = accuracy_score(y_test, predictions)
    roc_auc = roc_auc_score(y_test, probabilities)

    model_results.append({
        "Model": name,
        "Accuracy": accuracy,
        "ROC-AUC": roc_auc
    })

    print("\n" + "="*60)
    print(name)
    print("="*60)
    print("Accuracy:", accuracy)
    print("ROC-AUC:", roc_auc)
    print(classification_report(y_test, predictions))

model_results_df = pd.DataFrame(model_results).sort_values("ROC-AUC", ascending=False)
display(model_results_df)


# ============================================================
# 17. CONFUSION MATRIX FOR BEST MODEL
# ============================================================

best_model_name = model_results_df.iloc[0]["Model"]
best_model = models[best_model_name]

best_predictions = best_model.predict(X_test)

cm = confusion_matrix(y_test, best_predictions)

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title(f"Confusion Matrix - {best_model_name}")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()


# ============================================================
# 18. FEATURE IMPORTANCE
# ============================================================

if best_model_name in ["Random Forest", "Gradient Boosting"]:
    importances = best_model.feature_importances_

    importance_df = pd.DataFrame({
        "Feature": features,
        "Importance": importances
    }).sort_values("Importance", ascending=False)

    fig = px.bar(
        importance_df,
        x="Importance",
        y="Feature",
        orientation="h",
        title=f"Feature Importance - {best_model_name}"
    )
    fig.show()

    display(importance_df)
else:
    coefficients = best_model.named_steps["model"].coef_[0]

    coef_df = pd.DataFrame({
        "Feature": features,
        "Coefficient": coefficients
    }).sort_values("Coefficient", ascending=False)

    fig = px.bar(
        coef_df,
        x="Coefficient",
        y="Feature",
        orientation="h",
        title="Logistic Regression Coefficients"
    )
    fig.show()

    display(coef_df)


# ============================================================
# 19. ACTUAL VS PREDICTED SIGNAL
# ============================================================

prediction_df = pd.DataFrame(index=X_test.index)
prediction_df["Actual"] = y_test
prediction_df["Predicted"] = best_predictions
prediction_df["Close"] = close_prices[target_stock].loc[X_test.index]

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=prediction_df.index,
    y=prediction_df["Close"],
    mode="lines",
    name=f"{target_stock} Close Price"
))

buy_signals = prediction_df[prediction_df["Predicted"] == 1]

fig.add_trace(go.Scatter(
    x=buy_signals.index,
    y=buy_signals["Close"],
    mode="markers",
    name="Predicted Positive Return",
    marker=dict(size=7)
))

fig.update_layout(
    title=f"{target_stock} Price with ML Positive Return Signals",
    xaxis_title="Date",
    yaxis_title="Adjusted Close Price"
)

fig.show()


# ============================================================
# 20. SIMPLE STRATEGY BACKTEST
# If model predicts positive return, hold stock next day
# ============================================================

backtest = pd.DataFrame(index=X_test.index)
backtest["Actual_Return"] = ml_data["Return"].loc[X_test.index]
backtest["Signal"] = best_predictions
backtest["Strategy_Return"] = backtest["Signal"].shift(1) * backtest["Actual_Return"]
backtest = backtest.dropna()

backtest["Buy_and_Hold"] = (1 + backtest["Actual_Return"]).cumprod()
backtest["ML_Strategy"] = (1 + backtest["Strategy_Return"]).cumprod()

fig = px.line(
    backtest,
    x=backtest.index,
    y=["Buy_and_Hold", "ML_Strategy"],
    title=f"Backtest: ML Strategy vs Buy-and-Hold for {target_stock}",
    labels={"value": "Portfolio Value", "index": "Date"}
)
fig.show()

print("Final Buy-and-Hold Return:", backtest["Buy_and_Hold"].iloc[-1] - 1)
print("Final ML Strategy Return:", backtest["ML_Strategy"].iloc[-1] - 1)


# ============================================================
# 21. AUTO-GENERATED INTERPRETATION TEXT
# ============================================================

print("\n" + "="*80)
print("PROJECT INTERPRETATION SUMMARY")
print("="*80)

best_sharpe_stock = summary["Sharpe Ratio"].idxmax()
highest_vol_stock = summary["Annualized Volatility"].idxmax()
highest_beta_stock = beta_df["Beta vs BIST100"].idxmax()

print(f"""
1. Performance Analysis:
The normalized price and cumulative return graphs show how each selected BIST stock performed relative to BIST100.
Among the selected stocks, {best_sharpe_stock} achieved the highest Sharpe Ratio, meaning it provided the strongest risk-adjusted return in the analyzed period.

2. Risk Analysis:
The most volatile stock was {highest_vol_stock}, based on annualized volatility.
The drawdown graph shows the maximum historical decline from previous peaks, which is important for understanding downside risk.

3. Market Sensitivity:
The stock with the highest beta against BIST100 was {highest_beta_stock}.
This means it was the most sensitive to market-wide movements among the selected stocks.

4. Correlation Analysis:
The correlation heatmap shows whether the selected stocks move together.
High correlation implies limited diversification benefit, while lower correlation indicates better portfolio diversification potential.

5. Hypothesis Testing:
A two-sample t-test was applied to compare each stock's average daily return with BIST100.
If p-value < 0.05, the null hypothesis is rejected, suggesting a statistically significant difference from the market benchmark.

6. Normality and Stationarity:
Jarque-Bera tests check whether returns are normally distributed.
ADF tests check whether return series are stationary, which is important for financial modeling.

7. Machine Learning:
The ML section predicts whether {target_stock}'s next-day return will be positive.
The best performing model was {best_model_name}, based on ROC-AUC.
Technical indicators such as RSI, MACD, lagged returns, rolling volatility, and market return were used as predictive features.

8. Strategy Backtest:
The final graph compares a simple ML-based trading strategy against buy-and-hold.
This does not prove profitability, but it shows whether predictive signals contain useful information.
""")
```
