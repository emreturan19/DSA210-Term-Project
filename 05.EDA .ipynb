```python
# =========================
# DSA210 - CORRECTED EDA CODE
# 6 Stocks + BIST100 Benchmark + Hypothesis Testing
# =========================

!pip install yfinance -q

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# -------------------------
# 1) CONFIGURATION
# -------------------------

SELECTED_STOCKS = [
    "ASELS.IS",
    "THYAO.IS",
    "BIMAS.IS",
    "GARAN.IS",
    "TUPRS.IS",
    "KCHOL.IS"
]

BENCHMARK = "XU100.IS"   # BIST100
START_DATE = "2020-01-01"
END_DATE = "2025-12-31"

ALL_TICKERS = SELECTED_STOCKS + [BENCHMARK]

# -------------------------
# 2) DATA COLLECTION
# -------------------------

raw_data = yf.download(
    ALL_TICKERS,
    start=START_DATE,
    end=END_DATE,
    auto_adjust=True,
    progress=False
)

prices = raw_data["Close"]

prices = prices.dropna(how="all")
prices = prices.ffill().dropna()

print("Price data shape:", prices.shape)
display(prices.head())
display(prices.tail())

# -------------------------
# 3) DAILY RETURNS
# -------------------------

daily_returns = prices.pct_change().dropna()

stock_returns = daily_returns[SELECTED_STOCKS]
benchmark_returns = daily_returns[BENCHMARK]

print("Daily returns shape:", daily_returns.shape)
display(daily_returns.head())

# -------------------------
# 4) DESCRIPTIVE STATISTICS
# -------------------------

desc_stats = stock_returns.describe().T

desc_stats["mean_return"] = stock_returns.mean()
desc_stats["volatility_std"] = stock_returns.std()
desc_stats["annualized_return"] = stock_returns.mean() * 252
desc_stats["annualized_volatility"] = stock_returns.std() * np.sqrt(252)
desc_stats["min_return"] = stock_returns.min()
desc_stats["max_return"] = stock_returns.max()

display(desc_stats)

benchmark_stats = benchmark_returns.describe()
print("BIST100 Benchmark Statistics")
display(benchmark_stats)

# -------------------------
# 5) MISSING VALUE CHECK
# -------------------------

missing_values = prices.isnull().sum()
print("Missing values after cleaning:")
display(missing_values)

# -------------------------
# 6) NORMALIZED PRICE PLOT
# -------------------------

normalized_prices = prices / prices.iloc[0] * 100

plt.figure(figsize=(14, 7))
for ticker in SELECTED_STOCKS:
    plt.plot(normalized_prices.index, normalized_prices[ticker], label=ticker)

plt.plot(
    normalized_prices.index,
    normalized_prices[BENCHMARK],
    label="BIST100 Benchmark",
    linewidth=3,
    linestyle="--"
)

plt.title("Normalized Price Performance of Selected Stocks vs BIST100")
plt.xlabel("Date")
plt.ylabel("Normalized Price Index (Start = 100)")
plt.legend()
plt.grid(True)
plt.show()

# -------------------------
# 7) CUMULATIVE RETURNS PLOT
# -------------------------

cumulative_returns = (1 + daily_returns).cumprod() - 1

plt.figure(figsize=(14, 7))
for ticker in SELECTED_STOCKS:
    plt.plot(cumulative_returns.index, cumulative_returns[ticker], label=ticker)

plt.plot(
    cumulative_returns.index,
    cumulative_returns[BENCHMARK],
    label="BIST100 Benchmark",
    linewidth=3,
    linestyle="--"
)

plt.title("Cumulative Returns of Selected Stocks vs BIST100")
plt.xlabel("Date")
plt.ylabel("Cumulative Return")
plt.legend()
plt.grid(True)
plt.show()

# -------------------------
# 8) DAILY RETURN DISTRIBUTIONS
# -------------------------

for ticker in SELECTED_STOCKS:
    plt.figure(figsize=(8, 5))
    plt.hist(stock_returns[ticker], bins=50, alpha=0.75)
    plt.title(f"Distribution of Daily Returns - {ticker}")
    plt.xlabel("Daily Return")
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.show()

# -------------------------
# 9) BOX PLOT OF RETURNS
# -------------------------

plt.figure(figsize=(12, 6))
plt.boxplot(
    [stock_returns[ticker] for ticker in SELECTED_STOCKS],
    labels=SELECTED_STOCKS
)
plt.title("Box Plot of Daily Returns for Selected Stocks")
plt.xlabel("Stock")
plt.ylabel("Daily Return")
plt.grid(True)
plt.show()

# -------------------------
# 10) CORRELATION MATRIX
# -------------------------

correlation_matrix = stock_returns.corr()

plt.figure(figsize=(10, 7))
plt.imshow(correlation_matrix, interpolation="nearest")
plt.colorbar(label="Correlation Coefficient")
plt.xticks(range(len(SELECTED_STOCKS)), SELECTED_STOCKS, rotation=45)
plt.yticks(range(len(SELECTED_STOCKS)), SELECTED_STOCKS)
plt.title("Correlation Matrix of Selected Stock Returns")

for i in range(len(SELECTED_STOCKS)):
    for j in range(len(SELECTED_STOCKS)):
        plt.text(j, i, f"{correlation_matrix.iloc[i, j]:.2f}",
                 ha="center", va="center")

plt.tight_layout()
plt.show()

display(correlation_matrix)

# -------------------------
# 11) STOCKS VS BIST100 RETURN COMPARISON
# -------------------------

comparison_table = pd.DataFrame({
    "Stock Mean Daily Return": stock_returns.mean(),
    "BIST100 Mean Daily Return": benchmark_returns.mean(),
    "Difference": stock_returns.mean() - benchmark_returns.mean(),
    "Stock Volatility": stock_returns.std(),
    "BIST100 Volatility": benchmark_returns.std()
})

display(comparison_table)

# -------------------------
# 12) HYPOTHESIS TESTING - PAIRED T-TEST
# -------------------------

"""
H0: The mean daily return of the stock is equal to the mean daily return of BIST100.
H1: The mean daily return of the stock is different from the mean daily return of BIST100.

A paired t-test is used because each stock's daily return and BIST100 daily return are observed on the same trading dates.
Therefore, the observations are dependent rather than independent.
"""

test_results = []

for ticker in SELECTED_STOCKS:
    paired_data = pd.concat(
        [stock_returns[ticker], benchmark_returns],
        axis=1
    ).dropna()

    paired_data.columns = [ticker, "BIST100"]

    t_stat, p_value = stats.ttest_rel(
        paired_data[ticker],
        paired_data["BIST100"]
    )

    alpha = 0.05

    if p_value < alpha:
        conclusion = "Reject H0"
        interpretation = "There is a statistically significant difference from BIST100."
    else:
        conclusion = "Fail to reject H0"
        interpretation = "There is no statistically significant difference from BIST100."

    test_results.append({
        "Stock": ticker,
        "Mean Stock Return": paired_data[ticker].mean(),
        "Mean BIST100 Return": paired_data["BIST100"].mean(),
        "Mean Difference": paired_data[ticker].mean() - paired_data["BIST100"].mean(),
        "T-statistic": t_stat,
        "P-value": p_value,
        "Conclusion": conclusion,
        "Interpretation": interpretation
    })

hypothesis_results = pd.DataFrame(test_results)
display(hypothesis_results)

# -------------------------
# 13) VISUALIZATION OF MEAN RETURNS
# -------------------------

mean_returns = pd.concat([
    stock_returns.mean(),
    pd.Series({BENCHMARK: benchmark_returns.mean()})
])

plt.figure(figsize=(10, 6))
plt.bar(mean_returns.index, mean_returns.values)
plt.title("Mean Daily Returns: Selected Stocks vs BIST100")
plt.xlabel("Asset")
plt.ylabel("Mean Daily Return")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.show()

# -------------------------
# 14) RISK-RETURN SCATTER PLOT
# -------------------------

risk_return = pd.DataFrame({
    "Mean Daily Return": stock_returns.mean(),
    "Volatility": stock_returns.std()
})

plt.figure(figsize=(9, 6))
plt.scatter(risk_return["Volatility"], risk_return["Mean Daily Return"])

for ticker in SELECTED_STOCKS:
    plt.text(
        risk_return.loc[ticker, "Volatility"],
        risk_return.loc[ticker, "Mean Daily Return"],
        ticker
    )

plt.title("Risk-Return Profile of Selected Stocks")
plt.xlabel("Volatility / Standard Deviation of Daily Returns")
plt.ylabel("Mean Daily Return")
plt.grid(True)
plt.show()

display(risk_return)

# -------------------------
# 15) AUTOMATED EDA INTERPRETATION TEXT
# -------------------------

print("EDA INTERPRETATION SUMMARY")
print("-" * 50)

print("""
This corrected EDA uses BIST100 as the official market benchmark instead of using the average return of selected stocks.
This is important because BIST100 represents the broader Turkish stock market and provides a more valid comparison point.

The normalized price and cumulative return graphs show how each selected stock performed relative to the market benchmark.
Stocks that stay above the BIST100 line indicate stronger relative performance, while stocks below the benchmark indicate weaker performance.

The return distribution plots show that daily stock returns are centered around small values but contain extreme positive and negative observations.
This suggests that the selected stocks experience volatility and occasional large market movements.

The box plots compare the spread and outliers of daily returns across stocks.
Stocks with wider boxes or longer whiskers have higher return variability and therefore higher risk.

The correlation matrix shows the degree of co-movement among selected stocks.
High positive correlations suggest that stocks tend to move together, while lower correlations may provide diversification benefits.

For hypothesis testing, paired t-tests are applied between each stock's daily return and BIST100 daily return.
A paired t-test is appropriate because both values are observed on the same trading days, meaning the samples are dependent.
The null hypothesis states that there is no difference between the stock's average daily return and BIST100's average daily return.
If the p-value is below 0.05, the result suggests a statistically significant difference from the benchmark.
""")
```
