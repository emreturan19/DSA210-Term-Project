```python
# ============================================================
# FINAL FULL PROJECT CODE
# DSA210 TERM PROJECT
# EDA + Hypothesis Testing + Regression + Machine Learning
# ============================================================

!pip install yfinance statsmodels scipy scikit-learn openpyxl -q

# ============================================================
# 1. IMPORT LIBRARIES
# ============================================================

import os
import warnings
warnings.filterwarnings("ignore")

import yfinance as yf
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import statsmodels.api as sm

from scipy.stats import ttest_rel, jarque_bera, pearsonr
from statsmodels.tsa.stattools import adfuller

from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    classification_report,
    confusion_matrix,
    roc_curve
)

# ============================================================
# 2. OUTPUT FOLDERS
# ============================================================

os.makedirs("outputs/figures", exist_ok=True)
os.makedirs("outputs/tables", exist_ok=True)

# ============================================================
# 3. VISUAL SETTINGS
# ============================================================

sns.set_theme(style="whitegrid")

plt.rcParams["figure.figsize"] = (15, 7)
plt.rcParams["axes.titlesize"] = 16
plt.rcParams["axes.labelsize"] = 13
plt.rcParams["xtick.labelsize"] = 11
plt.rcParams["ytick.labelsize"] = 11

# ============================================================
# 4. PROJECT SETTINGS
# ============================================================

start_date = "2018-01-01"
end_date = "2025-12-31"

assets = {
    "THYAO": ["THYAO.IS"],
    "GARAN": ["GARAN.IS"],
    "ASELS": ["ASELS.IS"],
    "KCHOL": ["KCHOL.IS"],
    "TUPRS": ["TUPRS.IS"],
    "BIMAS": ["BIMAS.IS"],

    # Real BIST100 benchmark
    "BIST100": ["XU100.IS", "^XU100"],

    # Global market
    "SP500": ["^GSPC"],

    # Exchange rate
    "USDTRY": ["USDTRY=X", "TRY=X"]
}

stock_names = ["THYAO", "GARAN", "ASELS", "KCHOL", "TUPRS", "BIMAS"]

# ============================================================
# 5. ROBUST DOWNLOAD FUNCTION
# ============================================================

def download_asset(asset_name, ticker_options, start, end):
    """
    Downloads one asset using fallback Yahoo Finance tickers.
    If one ticker fails, the next alternative is tried.
    """

    if isinstance(ticker_options, str):
        ticker_options = [ticker_options]

    for ticker in ticker_options:
        print(f"Trying {asset_name}: {ticker}")

        try:
            df = yf.download(
                ticker,
                start=start,
                end=end,
                auto_adjust=True,
                progress=False,
                threads=False
            )

            if df is None or df.empty:
                print(f"  No data found for {ticker}")
                continue

            if "Close" not in df.columns:
                print(f"  Close column not found for {ticker}")
                continue

            output = pd.DataFrame(index=df.index)
            output[f"{asset_name}_Close"] = df["Close"]

            if "Volume" in df.columns:
                output[f"{asset_name}_Volume"] = df["Volume"]

            print(f"  SUCCESS: {asset_name} downloaded using {ticker}")
            return output

        except Exception as error:
            print(f"  Failed for {ticker}: {error}")

    raise ValueError(f"No usable data downloaded for {asset_name}")

# ============================================================
# 6. DOWNLOAD ALL DATA
# ============================================================

downloaded_frames = []

for asset_name, ticker_options in assets.items():
    downloaded_frames.append(
        download_asset(
            asset_name=asset_name,
            ticker_options=ticker_options,
            start=start_date,
            end=end_date
        )
    )

raw_data = pd.concat(downloaded_frames, axis=1).sort_index()

print("\nRaw data shape:", raw_data.shape)
display(raw_data.head())

# ============================================================
# 7. BUILD PRICE DATASET
# ============================================================

price_columns = [f"{asset}_Close" for asset in assets.keys()]
prices = raw_data[price_columns].copy()
prices.columns = list(assets.keys())

prices = prices.ffill().dropna()

print("\nCleaned prices shape:", prices.shape)
display(prices.head())

# ============================================================
# 8. BUILD VOLUME DATASET
# ============================================================

volume_columns = []

for stock in stock_names:
    col = f"{stock}_Volume"
    if col in raw_data.columns:
        volume_columns.append(col)

volume_data = raw_data[volume_columns].copy()

if not volume_data.empty:
    volume_data.columns = stock_names[:len(volume_columns)]
    volume_data = volume_data.reindex(prices.index).ffill()
else:
    print("WARNING: No stock volume data was downloaded.")

print("\nVolume data shape:", volume_data.shape)
display(volume_data.head())

# ============================================================
# 9. CALCULATE RETURNS
# ============================================================

returns = prices.pct_change()
returns = returns.replace([np.inf, -np.inf], np.nan)
returns = returns.dropna()

stock_returns = returns[stock_names].copy()

portfolio_return = stock_returns.mean(axis=1)

analysis_df = pd.DataFrame(index=returns.index)
analysis_df["Portfolio_Return"] = portfolio_return
analysis_df["BIST100_Return"] = returns["BIST100"]
analysis_df["SP500_Return"] = returns["SP500"]
analysis_df["USDTRY_Return"] = returns["USDTRY"]

analysis_df = analysis_df.dropna()
stock_returns = stock_returns.loc[analysis_df.index]

if not volume_data.empty:
    volume_data = volume_data.loc[analysis_df.index]
    average_volume = volume_data.mean(axis=1)
else:
    average_volume = pd.Series(index=analysis_df.index, dtype=float)

print("\nFinal analysis dataset shape:", analysis_df.shape)
display(analysis_df.head())

prices.to_csv("outputs/tables/clean_prices.csv")
returns.to_csv("outputs/tables/daily_returns.csv")
analysis_df.to_csv("outputs/tables/analysis_dataset.csv")

# ============================================================
# 10. DESCRIPTIVE STATISTICS
# ============================================================

desc_table = pd.DataFrame(index=stock_names)

desc_table["Mean Daily Return"] = stock_returns.mean()
desc_table["Daily Volatility"] = stock_returns.std()
desc_table["Annualized Return"] = stock_returns.mean() * 252
desc_table["Annualized Volatility"] = stock_returns.std() * np.sqrt(252)
desc_table["Skewness"] = stock_returns.skew()
desc_table["Kurtosis"] = stock_returns.kurtosis()

for stock in stock_names:
    cumulative_stock = (1 + stock_returns[stock]).cumprod()
    running_max_stock = cumulative_stock.cummax()
    stock_drawdown = (cumulative_stock - running_max_stock) / running_max_stock

    desc_table.loc[stock, "Maximum Drawdown"] = stock_drawdown.min()

    common_beta = pd.concat(
        [stock_returns[stock], analysis_df["BIST100_Return"]],
        axis=1
    ).dropna()

    common_beta.columns = ["Stock_Return", "BIST100_Return"]

    beta = (
        common_beta["Stock_Return"].cov(common_beta["BIST100_Return"])
        /
        common_beta["BIST100_Return"].var()
    )

    desc_table.loc[stock, "Beta vs BIST100"] = beta

desc_table = desc_table.round(6)

display(desc_table)

desc_table.to_csv("outputs/tables/descriptive_statistics.csv")

# ============================================================
# 11. FIGURE 1 - CUMULATIVE RETURNS
# ============================================================

cumulative_portfolio = (1 + analysis_df["Portfolio_Return"]).cumprod()
cumulative_bist100 = (1 + analysis_df["BIST100_Return"]).cumprod()
cumulative_sp500 = (1 + analysis_df["SP500_Return"]).cumprod()

plt.figure(figsize=(15, 7))

plt.plot(cumulative_portfolio, label="Equal-Weighted Portfolio", linewidth=2.5)
plt.plot(cumulative_bist100, label="BIST100 Benchmark", linewidth=2.5)
plt.plot(cumulative_sp500, label="S&P500", linewidth=2, alpha=0.8)

plt.title("Figure 1: Cumulative Return Comparison")
plt.xlabel("Date")
plt.ylabel("Cumulative Growth of 1 Unit")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()

plt.savefig("outputs/figures/fig1_cumulative_returns.png", dpi=300, bbox_inches="tight")
plt.show()

# ============================================================
# 12. FIGURE 2 - PORTFOLIO RETURN DISTRIBUTION
# ============================================================

portfolio_skew = analysis_df["Portfolio_Return"].skew()
portfolio_kurt = analysis_df["Portfolio_Return"].kurtosis()

plt.figure(figsize=(15, 7))

sns.histplot(
    analysis_df["Portfolio_Return"],
    bins=80,
    kde=True
)

plt.axvline(
    analysis_df["Portfolio_Return"].mean(),
    linestyle="--",
    linewidth=2,
    label="Mean Return"
)

plt.title(
    f"Figure 2: Portfolio Daily Return Distribution | Skewness = {portfolio_skew:.2f}, Kurtosis = {portfolio_kurt:.2f}"
)

plt.xlabel("Daily Portfolio Return")
plt.ylabel("Frequency")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()

plt.savefig("outputs/figures/fig2_return_distribution.png", dpi=300, bbox_inches="tight")
plt.show()

# ============================================================
# 13. FIGURE 3 - STOCK RETURN BOXPLOT
# ============================================================

plt.figure(figsize=(15, 7))

sns.boxplot(data=stock_returns)

plt.title("Figure 3: Daily Return Distribution and Outliers by Stock")
plt.xlabel("Stock")
plt.ylabel("Daily Return")
plt.xticks(rotation=45)
plt.grid(alpha=0.3)
plt.tight_layout()

plt.savefig("outputs/figures/fig3_stock_boxplot.png", dpi=300, bbox_inches="tight")
plt.show()

# ============================================================
# 14. FIGURE 4 - CORRELATION HEATMAP
# ============================================================

corr_df = pd.concat(
    [
        stock_returns,
        analysis_df[["Portfolio_Return", "BIST100_Return", "SP500_Return", "USDTRY_Return"]]
    ],
    axis=1
).dropna()

plt.figure(figsize=(15, 11))

sns.heatmap(
    corr_df.corr(),
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    center=0,
    linewidths=0.5
)

plt.title("Figure 4: Correlation Matrix of Stocks, Portfolio, Market and Macro Variables")
plt.xlabel("Variables")
plt.ylabel("Variables")
plt.tight_layout()

plt.savefig("outputs/figures/fig4_correlation_heatmap.png", dpi=300, bbox_inches="tight")
plt.show()

# ============================================================
# 15. FIGURE 5 - ROLLING VOLATILITY
# ============================================================

rolling_portfolio_vol = analysis_df["Portfolio_Return"].rolling(30).std()
rolling_bist100_vol = analysis_df["BIST100_Return"].rolling(30).std()

plt.figure(figsize=(15, 7))

plt.plot(
    rolling_portfolio_vol,
    label="Portfolio 30-Day Rolling Volatility",
    linewidth=2
)

plt.plot(
    rolling_bist100_vol,
    label="BIST100 30-Day Rolling Volatility",
    linewidth=2
)

plt.title("Figure 5: Rolling Volatility Comparison")
plt.xlabel("Date")
plt.ylabel("30-Day Rolling Volatility")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()

plt.savefig("outputs/figures/fig5_rolling_volatility.png", dpi=300, bbox_inches="tight")
plt.show()

# ============================================================
# 16. FIGURE 6 - ROLLING BETA
# ============================================================

rolling_beta = (
    analysis_df["Portfolio_Return"].rolling(60).cov(analysis_df["BIST100_Return"])
    /
    analysis_df["BIST100_Return"].rolling(60).var()
)

rolling_beta = rolling_beta.dropna()

plt.figure(figsize=(15, 7))

plt.plot(
    rolling_beta,
    label="60-Day Rolling Beta vs BIST100",
    linewidth=2.2
)

plt.axhline(
    1,
    linestyle="--",
    linewidth=1.8,
    label="Beta = 1"
)

plt.axhline(
    rolling_beta.mean(),
    linestyle="--",
    linewidth=1.8,
    label=f"Average Beta = {rolling_beta.mean():.2f}"
)

plt.title("Figure 6: Portfolio Rolling Beta Against BIST100")
plt.xlabel("Date")
plt.ylabel("Rolling Beta")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()

plt.savefig("outputs/figures/fig6_rolling_beta.png", dpi=300, bbox_inches="tight")
plt.show()

# ============================================================
# 17. FIGURE 7 - DRAWDOWN ANALYSIS
# ============================================================

running_max = cumulative_portfolio.cummax()
drawdown = (cumulative_portfolio - running_max) / running_max

plt.figure(figsize=(15, 7))

plt.plot(drawdown, linewidth=2)
plt.fill_between(drawdown.index, drawdown, 0, alpha=0.25)

plt.title(f"Figure 7: Portfolio Drawdown Analysis | Maximum Drawdown = {drawdown.min():.2%}")
plt.xlabel("Date")
plt.ylabel("Drawdown")
plt.grid(alpha=0.3)
plt.tight_layout()

plt.savefig("outputs/figures/fig7_drawdown.png", dpi=300, bbox_inches="tight")
plt.show()

# ============================================================
# 18. FIGURE 8 - VOLUME VS VOLATILITY
# ============================================================

if average_volume.notna().sum() > 10:
    volume_volatility_df = pd.DataFrame(index=analysis_df.index)
    volume_volatility_df["Average_Volume"] = average_volume
    volume_volatility_df["Absolute_Portfolio_Return"] = analysis_df["Portfolio_Return"].abs()
    volume_volatility_df = volume_volatility_df.dropna()

    corr_volume_volatility, p_volume_volatility = pearsonr(
        volume_volatility_df["Average_Volume"],
        volume_volatility_df["Absolute_Portfolio_Return"]
    )

    plt.figure(figsize=(15, 7))

    plt.scatter(
        volume_volatility_df["Average_Volume"],
        volume_volatility_df["Absolute_Portfolio_Return"],
        alpha=0.35
    )

    plt.title(
        f"Figure 8: Trading Volume vs Absolute Portfolio Return | r = {corr_volume_volatility:.3f}, p = {p_volume_volatility:.4f}"
    )

    plt.xlabel("Average Trading Volume")
    plt.ylabel("Absolute Portfolio Return")
    plt.grid(alpha=0.3)
    plt.tight_layout()

    plt.savefig("outputs/figures/fig8_volume_vs_volatility.png", dpi=300, bbox_inches="tight")
    plt.show()

else:
    corr_volume_volatility = np.nan
    p_volume_volatility = np.nan
    print("Volume data is not available enough. Figure 8 skipped.")

# ============================================================
# 19. HYPOTHESIS TEST 1 - PAIRED T-TEST
# ============================================================

paired_results = []

for stock in stock_names:
    common = pd.concat(
        [
            stock_returns[stock],
            analysis_df["BIST100_Return"]
        ],
        axis=1
    ).dropna()

    common.columns = ["Stock_Return", "BIST100_Return"]

    t_stat, p_value = ttest_rel(
        common["Stock_Return"],
        common["BIST100_Return"]
    )

    decision = "Reject H0" if p_value < 0.05 else "Fail to Reject H0"

    paired_results.append(
        [
            stock,
            common["Stock_Return"].mean(),
            common["BIST100_Return"].mean(),
            common["Stock_Return"].mean() - common["BIST100_Return"].mean(),
            t_stat,
            p_value,
            decision
        ]
    )

paired_test_df = pd.DataFrame(
    paired_results,
    columns=[
        "Stock",
        "Mean Stock Return",
        "Mean BIST100 Return",
        "Mean Difference",
        "t-statistic",
        "p-value",
        "Decision"
    ]
)

paired_test_df = paired_test_df.round(6)

display(paired_test_df)

paired_test_df.to_csv("outputs/tables/paired_t_test_results.csv", index=False)

print("""
PAIRED T-TEST INTERPRETATION

H0: Mean stock return is equal to mean BIST100 return.
HA: Mean stock return is different from mean BIST100 return.

A paired t-test is used because each stock return and the BIST100 return are observed on the same trading dates.
This means the observations are dependent rather than independent.

Decision rule:
If p-value < 0.05, reject H0.
If p-value >= 0.05, fail to reject H0.
""")

# ============================================================
# 20. HYPOTHESIS TEST 2 - JARQUE-BERA NORMALITY TEST
# ============================================================

jb_results = []

for stock in stock_names:
    jb_stat, jb_p = jarque_bera(stock_returns[stock].dropna())

    decision = "Reject Normality" if jb_p < 0.05 else "Fail to Reject Normality"

    jb_results.append(
        [
            stock,
            jb_stat,
            jb_p,
            stock_returns[stock].skew(),
            stock_returns[stock].kurtosis(),
            decision
        ]
    )

jb_df = pd.DataFrame(
    jb_results,
    columns=[
        "Stock",
        "JB Statistic",
        "p-value",
        "Skewness",
        "Kurtosis",
        "Decision"
    ]
)

jb_df = jb_df.round(6)

display(jb_df)

jb_df.to_csv("outputs/tables/jarque_bera_results.csv", index=False)

print("""
JARQUE-BERA TEST INTERPRETATION

H0: Returns are normally distributed.
HA: Returns are not normally distributed.

Small p-values indicate rejection of normality.
This is common in financial returns due to skewness, heavy tails and extreme observations.
""")

# ============================================================
# 21. HYPOTHESIS TEST 3 - ADF STATIONARITY TEST
# ============================================================

adf_results = []

for stock in stock_names:
    adf_stat, adf_p, _, _, critical_values, _ = adfuller(stock_returns[stock].dropna())

    decision = "Stationary" if adf_p < 0.05 else "Non-stationary"

    adf_results.append(
        [
            stock,
            adf_stat,
            adf_p,
            critical_values["1%"],
            critical_values["5%"],
            critical_values["10%"],
            decision
        ]
    )

adf_df = pd.DataFrame(
    adf_results,
    columns=[
        "Stock",
        "ADF Statistic",
        "p-value",
        "Critical Value 1%",
        "Critical Value 5%",
        "Critical Value 10%",
        "Decision"
    ]
)

adf_df = adf_df.round(6)

display(adf_df)

adf_df.to_csv("outputs/tables/adf_results.csv", index=False)

print("""
ADF TEST INTERPRETATION

H0: The return series is non-stationary.
HA: The return series is stationary.

If p-value < 0.05, the return series is considered stationary.
Stationarity is important for time-series modeling and machine learning reliability.
""")

# ============================================================
# 22. REGRESSION ANALYSIS
# ============================================================

regression_df = analysis_df[
    [
        "Portfolio_Return",
        "BIST100_Return",
        "SP500_Return",
        "USDTRY_Return"
    ]
].dropna()

X = regression_df[
    [
        "BIST100_Return",
        "SP500_Return",
        "USDTRY_Return"
    ]
]

X = sm.add_constant(X)
y = regression_df["Portfolio_Return"]

ols_model = sm.OLS(y, X).fit()

print(ols_model.summary())

coef_table = pd.DataFrame(
    {
        "Coefficient": ols_model.params,
        "p-value": ols_model.pvalues,
        "Significant at 5%": ols_model.pvalues < 0.05
    }
)

coef_table = coef_table.round(6)

display(coef_table)

coef_table.to_csv("outputs/tables/regression_coefficients.csv")

plt.figure(figsize=(12, 7))

coef_table.loc[
    [
        "BIST100_Return",
        "SP500_Return",
        "USDTRY_Return"
    ],
    "Coefficient"
].plot(kind="bar")

plt.title("Figure 9: Regression Coefficients for Portfolio Return Model")
plt.xlabel("Explanatory Variable")
plt.ylabel("Regression Coefficient")
plt.xticks(rotation=30)
plt.grid(alpha=0.3)
plt.tight_layout()

plt.savefig("outputs/figures/fig9_regression_coefficients.png", dpi=300, bbox_inches="tight")
plt.show()

print("""
REGRESSION INTERPRETATION

Dependent variable:
Portfolio_Return

Independent variables:
BIST100_Return, SP500_Return, USDTRY_Return

A larger and statistically significant BIST100 coefficient indicates that domestic market movements
are the dominant explanatory factor for the portfolio.
""")

# ============================================================
# 23. MACHINE LEARNING FEATURE ENGINEERING
# ============================================================

ml_df = analysis_df.copy()

ml_df["Lag_1"] = ml_df["Portfolio_Return"].shift(1)
ml_df["Lag_2"] = ml_df["Portfolio_Return"].shift(2)
ml_df["Lag_3"] = ml_df["Portfolio_Return"].shift(3)
ml_df["Lag_5"] = ml_df["Portfolio_Return"].shift(5)

ml_df["Rolling_Return_5"] = ml_df["Portfolio_Return"].rolling(5).mean()
ml_df["Rolling_Return_10"] = ml_df["Portfolio_Return"].rolling(10).mean()
ml_df["Rolling_Volatility_5"] = ml_df["Portfolio_Return"].rolling(5).std()
ml_df["Rolling_Volatility_10"] = ml_df["Portfolio_Return"].rolling(10).std()

ml_df["BIST100_Lag_1"] = ml_df["BIST100_Return"].shift(1)
ml_df["SP500_Lag_1"] = ml_df["SP500_Return"].shift(1)
ml_df["USDTRY_Lag_1"] = ml_df["USDTRY_Return"].shift(1)

# Target: next-day portfolio return direction
ml_df["Target"] = (ml_df["Portfolio_Return"].shift(-1) > 0).astype(int)

ml_df = ml_df.dropna()

feature_columns = [
    "BIST100_Return",
    "SP500_Return",
    "USDTRY_Return",
    "Lag_1",
    "Lag_2",
    "Lag_3",
    "Lag_5",
    "Rolling_Return_5",
    "Rolling_Return_10",
    "Rolling_Volatility_5",
    "Rolling_Volatility_10",
    "BIST100_Lag_1",
    "SP500_Lag_1",
    "USDTRY_Lag_1"
]

X_ml = ml_df[feature_columns]
y_ml = ml_df["Target"]

split_index = int(len(ml_df) * 0.80)

X_train = X_ml.iloc[:split_index]
X_test = X_ml.iloc[split_index:]

y_train = y_ml.iloc[:split_index]
y_test = y_ml.iloc[split_index:]

print("\nML dataset shape:", ml_df.shape)
print("Training size:", X_train.shape)
print("Test size:", X_test.shape)

ml_df.to_csv("outputs/tables/ml_dataset.csv")

# ============================================================
# 24. MACHINE LEARNING MODELS
# ============================================================

models = {
    "Logistic Regression": Pipeline(
        steps=[
            ("scaler", StandardScaler()),
            ("model", LogisticRegression(max_iter=1000))
        ]
    ),

    "Random Forest": RandomForestClassifier(
        n_estimators=300,
        max_depth=6,
        min_samples_leaf=5,
        random_state=42
    ),

    "Gradient Boosting": GradientBoostingClassifier(
        n_estimators=200,
        learning_rate=0.05,
        max_depth=3,
        random_state=42
    )
}

ml_results = []
predictions = {}
probabilities = {}

for model_name, model in models.items():

    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    if hasattr(model, "predict_proba"):
        y_prob = model.predict_proba(X_test)[:, 1]
    else:
        y_prob = y_pred

    predictions[model_name] = y_pred
    probabilities[model_name] = y_prob

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, zero_division=0)
    recall = recall_score(y_test, y_pred, zero_division=0)
    f1 = f1_score(y_test, y_pred, zero_division=0)
    roc_auc = roc_auc_score(y_test, y_prob)

    ml_results.append(
        [
            model_name,
            accuracy,
            precision,
            recall,
            f1,
            roc_auc
        ]
    )

    print("\n" + "=" * 70)
    print(model_name)
    print("=" * 70)
    print(classification_report(y_test, y_pred, zero_division=0))

ml_results_df = pd.DataFrame(
    ml_results,
    columns=[
        "Model",
        "Accuracy",
        "Precision",
        "Recall",
        "F1-score",
        "ROC-AUC"
    ]
)

ml_results_df = ml_results_df.round(6)

display(ml_results_df)

ml_results_df.to_csv("outputs/tables/ml_model_performance.csv", index=False)

# ============================================================
# 25. FIGURE 10 - MODEL ACCURACY COMPARISON
# ============================================================

plt.figure(figsize=(10, 6))

sns.barplot(
    data=ml_results_df,
    x="Model",
    y="Accuracy"
)

plt.title("Figure 10: Model Accuracy Comparison")
plt.xlabel("Machine Learning Model")
plt.ylabel("Accuracy")
plt.ylim(0, 1)
plt.xticks(rotation=20)
plt.grid(alpha=0.3)
plt.tight_layout()

plt.savefig("outputs/figures/fig10_model_accuracy_comparison.png", dpi=300, bbox_inches="tight")
plt.show()

# ============================================================
# 26. FIGURE 11 - ROC-AUC COMPARISON
# ============================================================

plt.figure(figsize=(10, 6))

sns.barplot(
    data=ml_results_df,
    x="Model",
    y="ROC-AUC"
)

plt.title("Figure 11: ROC-AUC Comparison")
plt.xlabel("Machine Learning Model")
plt.ylabel("ROC-AUC")
plt.ylim(0, 1)
plt.xticks(rotation=20)
plt.grid(alpha=0.3)
plt.tight_layout()

plt.savefig("outputs/figures/fig11_roc_auc_comparison.png", dpi=300, bbox_inches="tight")
plt.show()

# ============================================================
# 27. FIGURE 12 - ROC CURVES
# ============================================================

plt.figure(figsize=(10, 7))

for model_name in models.keys():
    fpr, tpr, _ = roc_curve(y_test, probabilities[model_name])

    auc_value = roc_auc_score(y_test, probabilities[model_name])

    plt.plot(
        fpr,
        tpr,
        linewidth=2,
        label=f"{model_name} (AUC = {auc_value:.3f})"
    )

plt.plot([0, 1], [0, 1], linestyle="--", linewidth=2, label="Random Classifier")

plt.title("Figure 12: ROC Curves of Machine Learning Models")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()

plt.savefig("outputs/figures/fig12_roc_curves.png", dpi=300, bbox_inches="tight")
plt.show()

# ============================================================
# 28. FIGURE 13 - CONFUSION MATRIX FOR BEST MODEL
# ============================================================

best_model_name = ml_results_df.sort_values(by="ROC-AUC", ascending=False).iloc[0]["Model"]

best_pred = predictions[best_model_name]

cm = confusion_matrix(y_test, best_pred)

plt.figure(figsize=(7, 6))

sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=["Negative Return", "Positive Return"],
    yticklabels=["Negative Return", "Positive Return"]
)

plt.title(f"Figure 13: Confusion Matrix - {best_model_name}")
plt.xlabel("Predicted Class")
plt.ylabel("Actual Class")
plt.tight_layout()

plt.savefig("outputs/figures/fig13_confusion_matrix_best_model.png", dpi=300, bbox_inches="tight")
plt.show()

# ============================================================
# 29. FIGURE 14 - FEATURE IMPORTANCE
# ============================================================

if best_model_name == "Random Forest":
    best_model = models["Random Forest"]
    importance_values = best_model.feature_importances_

elif best_model_name == "Gradient Boosting":
    best_model = models["Gradient Boosting"]
    importance_values = best_model.feature_importances_

else:
    logistic_model = models["Logistic Regression"].named_steps["model"]
    importance_values = np.abs(logistic_model.coef_[0])

importance_df = pd.DataFrame(
    {
        "Feature": feature_columns,
        "Importance": importance_values
    }
).sort_values(by="Importance", ascending=False)

display(importance_df.round(6))

importance_df.to_csv("outputs/tables/ml_feature_importance.csv", index=False)

plt.figure(figsize=(12, 7))

sns.barplot(
    data=importance_df.head(12),
    x="Importance",
    y="Feature"
)

plt.title(f"Figure 14: Feature Importance - {best_model_name}")
plt.xlabel("Importance Score")
plt.ylabel("Feature")
plt.grid(alpha=0.3)
plt.tight_layout()

plt.savefig("outputs/figures/fig14_feature_importance.png", dpi=300, bbox_inches="tight")
plt.show()

# ============================================================
# 30. ACTUAL CLASS VS PREDICTED CLASS
# ============================================================

comparison_df = pd.DataFrame(index=y_test.index)
comparison_df["Actual_Class"] = y_test
comparison_df["Predicted_Class"] = best_pred
comparison_df["Predicted_Probability"] = probabilities[best_model_name]

comparison_df.to_csv("outputs/tables/ml_actual_vs_predicted.csv")

plt.figure(figsize=(15, 6))

plt.plot(
    comparison_df.index,
    comparison_df["Actual_Class"],
    label="Actual Direction",
    alpha=0.8
)

plt.plot(
    comparison_df.index,
    comparison_df["Predicted_Probability"],
    label=f"Predicted Probability - {best_model_name}",
    alpha=0.8
)

plt.title(f"Figure 15: Actual Return Direction vs Predicted Probability - {best_model_name}")
plt.xlabel("Date")
plt.ylabel("Class / Predicted Probability")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()

plt.savefig("outputs/figures/fig15_actual_vs_predicted_probability.png", dpi=300, bbox_inches="tight")
plt.show()

# ============================================================
# 31. FINAL SUMMARY TABLE
# ============================================================

summary = {
    "Average Portfolio Return": analysis_df["Portfolio_Return"].mean(),
    "Portfolio Volatility": analysis_df["Portfolio_Return"].std(),
    "Portfolio Skewness": analysis_df["Portfolio_Return"].skew(),
    "Portfolio Kurtosis": analysis_df["Portfolio_Return"].kurtosis(),
    "Maximum Drawdown": drawdown.min(),
    "Average Rolling Beta": rolling_beta.mean(),
    "Volume-Volatility Correlation": corr_volume_volatility,
    "Volume-Volatility p-value": p_volume_volatility,
    "Regression R-squared": ols_model.rsquared,
    "Best ML Model": best_model_name,
    "Best ML ROC-AUC": ml_results_df.sort_values(by="ROC-AUC", ascending=False).iloc[0]["ROC-AUC"],
    "Best ML Accuracy": ml_results_df.sort_values(by="ROC-AUC", ascending=False).iloc[0]["Accuracy"]
}

summary_df = pd.DataFrame(summary, index=["Value"]).T

display(summary_df)

summary_df.to_csv("outputs/tables/final_summary.csv")

print("""
FULL PIPELINE COMPLETED SUCCESSFULLY.

Saved figures:
outputs/figures/

Saved tables:
outputs/tables/

Important outputs:
- descriptive_statistics.csv
- paired_t_test_results.csv
- jarque_bera_results.csv
- adf_results.csv
- regression_coefficients.csv
- ml_model_performance.csv
- ml_feature_importance.csv
- final_summary.csv
""")
```
