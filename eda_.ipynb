Kodu düzenlemek için hücreyi tıklamanız yeterlidir. Sonrasında düzenlemeye başlayabilirsiniz.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.gridspec as gridspec

# ═══════════════════════════════════════════════════════════════
# AYARLAR VE PALET (Eksikse kod çalışmaz)
# ═══════════════════════════════════════════════════════════════
PALETTE = {
    'bg': '#121212',
    'card': '#1E1E26',
    'text': '#E0E0E0',
    'subtext': '#A0A0A0',
    'grid': '#303033',
    'accent1': '#00F5D4',  # Turkuaz
    'accent2': '#F15BB5',  # Pembe/Kırmızı
    'accent3': '#00BBF9'   # Mavi
}

STOCK_COLORS = ['#00F5D4', '#F15BB5', '#00BBF9', '#FEE440', '#9b5de5', '#f1faee']

# ═══════════════════════════════════════════════════════════════
# GRAFİK 1 – ÖZET İSTATİSTİKLER (Isı Haritası)
# ═══════════════════════════════════════════════════════════════
def plot_summary_stats(data):
    if data.empty:
        print("Grafik 1 atlandı: Veri boş.")
        return

    metrics = data.columns
    fig, ax = plt.subplots(figsize=(12, len(data) * 0.8), facecolor=PALETTE['bg'])
    ax.set_facecolor(PALETTE['card'])

    # Normalize per column for coloring
    normed = (data - data.min()) / (data.max() - data.min() + 1e-9)

    for i, stock in enumerate(data.index):
        for j, col in enumerate(metrics):
            val   = data.loc[stock, col]
            nval  = normed.loc[stock, col]
            color = plt.cm.RdYlGn(nval)
            ax.add_patch(plt.Rectangle((j, i), 1, 1, color=color, alpha=0.8))
            ax.text(j + 0.5, i + 0.5, f'{val:.4f}',
                    ha='center', va='center', fontsize=11,
                    color='white', fontweight='bold')

    ax.set_xlim(0, len(metrics))
    ax.set_ylim(0, len(data))
    ax.set_xticks([x + 0.5 for x in range(len(metrics))])
    ax.set_xticklabels(metrics, fontsize=12, fontweight='bold', color=PALETTE['text'])
    ax.set_yticks([y + 0.5 for y in range(len(data))])
    ax.set_yticklabels(data.index, fontsize=12, color=PALETTE['text'])
    ax.set_title('Özet İstatistikler – Günlük Getiriler', fontsize=16,
                 fontweight='bold', color=PALETTE['text'], pad=15)
    ax.grid(False)

    fig.tight_layout()
    fig.savefig('grafik1_ozet_istatistikler.png', dpi=150,
                bbox_inches='tight', facecolor=PALETTE['bg'])
    print("✓ grafik1_ozet_istatistikler.png")
    plt.close()

# ═══════════════════════════════════════════════════════════════
# GRAFİK 5 – KORELASYON MATRİSİ (Hata Veren Kısım Düzeltildi)
# ═══════════════════════════════════════════════════════════════
def plot_correlation(stock_ret, bist_ret, sp500_ret, usdtry_ret, port_ret):
    # Veri kontrolü ve birleştirme
    try:
        all_ret = stock_ret.copy()
        all_ret['BIST100'] = bist_ret
        all_ret['S&P500']  = sp500_ret
        all_ret['USD/TRY'] = usdtry_ret
        all_ret['Portföy'] = port_ret
        corr = all_ret.corr()
    except Exception as e:
        print(f"Grafik 5 atlandı: Veri hatası -> {e}")
        return

    fig, ax = plt.subplots(figsize=(12, 10), facecolor=PALETTE['bg'])
    mask = np.triu(np.ones_like(corr, dtype=bool), k=1)

    cmap = sns.diverging_palette(220, 20, as_cmap=True)
    sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
                annot=True, fmt='.2f', annot_kws={'size': 11, 'weight': 'bold'},
                linewidths=0.5, linecolor=PALETTE['bg'],
                square=True, ax=ax,
                cbar_kws={'shrink': 0.8, 'label': 'Korelasyon'})

    ax.set_title('Korelasyon Matrisi – Varlıklar & Makro Veriler',
                 fontsize=15, fontweight='bold', color=PALETTE['text'], pad=15)

    plt.xticks(rotation=45, color=PALETTE['text'])
    plt.yticks(rotation=0, color=PALETTE['text'])

    fig.tight_layout()
    fig.savefig('grafik5_korelasyon.png', dpi=150,
                bbox_inches='tight', facecolor=PALETTE['bg'])
    print("✓ grafik5_korelasyon.png")
    plt.close()

# ═══════════════════════════════════════════════════════════════
# GRAFİK 7 – REGRESYON (İndentasyon Hatası Çözüldü)
# ═══════════════════════════════════════════════════════════════
def plot_regression(stock_ret, bist_ret, sp500_ret, usdtry_ret, port_ret):
    if bist_ret.empty or sp500_ret.empty or usdtry_ret.empty:
        print("Grafik 7 atlandı: Eksik veri.")
        return

    X = pd.DataFrame({
        'BIST100': bist_ret,
        'S&P500':  sp500_ret,
        'USD/TRY': usdtry_ret,
    })

    # Hata veren nokta burasıydı, parantez içine alarak güvenli hale getirdik
    common = (X.dropna().index
              .intersection(port_ret.dropna().index))

    if len(common) < 10:
        print("Grafik 7 atlandı: Yetersiz ortak veri.")
        return

    X_c = X.loc[common]
    y_c = port_ret.loc[common]

    # Regresyon Hesaplama
    X_mat = np.column_stack([np.ones(len(X_c)), X_c.values])
    coefs, _, _, _ = np.linalg.lstsq(X_mat, y_c.values, rcond=None)

    labels_reg = ['Sabit (β₀)', 'BIST100 (β₁)', 'S&P500 (β₂)', 'USD/TRY (β₃)']
    colors_reg = [PALETTE['subtext'], PALETTE['accent1'], PALETTE['accent3'], PALETTE['accent2']]

    fig, ax = plt.subplots(figsize=(10, 6), facecolor=PALETTE['bg'])
    ax.set_facecolor(PALETTE['card'])

    y_pos = np.arange(len(coefs))
    ax.barh(y_pos, coefs, color=colors_reg, alpha=0.8)

    for i, v in enumerate(coefs):
        ax.text(v, i, f' {v:.5f}', va='center', color='white', fontweight='bold')

    ax.set_yticks(y_pos)
    ax.set_yticklabels(labels_reg, color=PALETTE['text'])
    ax.set_title('Portföy Duyarlılık Analizi (Regresyon Katsayıları)', color=PALETTE['text'], fontsize=14)

    plt.tight_layout()
    fig.savefig('grafik7_regresyon.png', dpi=150, facecolor=PALETTE['bg'])
    print("✓ grafik7_regresyon.png")
    plt.close()

# ═══════════════════════════════════════════════════════════════
# ÇALIŞTIRMA ÖRNEĞİ
# ═══════════════════════════════════════════════════════════════
if __name__ == '__main__':
    print("Grafikler oluşturuluyor...")
    # Burada senin hazırladığın veri çerçevelerini (stock_ret, vb.) fonksiyona göndermen lazım.
    # Örn: plot_correlation(stock_ret, bist_ret, sp500_ret, usdtry_ret, port_ret)
```

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import matplotlib.gridspec as gridspec

# 1. GÖRSEL AYARLAR VE RENK PALETİ
PALETTE = {
    'bg': '#121212', 'card': '#1E1E26', 'text': '#E0E0E0',
    'subtext': '#A0A0A0', 'grid': '#303033',
    'accent1': '#00F5D4', 'accent2': '#F15BB5', 'accent3': '#00BBF9'
}
STOCK_COLORS = ['#00F5D4', '#F15BB5', '#00BBF9', '#FEE440', '#9b5de5']
plt.rcParams['axes.facecolor'] = PALETTE['card']
plt.rcParams['figure.facecolor'] = PALETTE['bg']
plt.rcParams['text.color'] = PALETTE['text']

# 2. ÖRNEK VERİ OLUŞTURMA (Hata almamak için gerekli)
np.random.seed(42)
dates = pd.date_range('2023-01-01', periods=252)
LABELS = ['Hisse_A', 'Hisse_B', 'Hisse_C']
stock_ret = pd.DataFrame(np.random.normal(0.001, 0.02, (252, 3)), index=dates, columns=LABELS)
volumes = pd.DataFrame(np.random.randint(100, 1000, (252, 3)), index=dates, columns=LABELS)
bist_ret = pd.Series(np.random.normal(0.0008, 0.015, 252), index=dates)
sp500_ret = pd.Series(np.random.normal(0.0005, 0.012, 252), index=dates)
usdtry_ret = pd.Series(np.random.normal(0.002, 0.008, 252), index=dates)
port_ret = stock_ret.mean(axis=1)

# 3. GRAFİK FONKSİYONLARI (Düzeltilmiş Hali)
def run_all_plots():
    # --- GRAFİK 1: KORELASYON ---
    all_data = stock_ret.copy()
    all_data['BIST100'], all_data['S&P500'], all_data['USD/TRY'], all_data['Portföy'] = bist_ret, sp500_ret, usdtry_ret, port_ret
    corr = all_data.corr()

    plt.figure(figsize=(10, 8))
    mask = np.triu(np.ones_like(corr, dtype=bool), k=1)
    sns.heatmap(corr, mask=mask, cmap='RdYlGn', annot=True, fmt='.2f', center=0)
    plt.title('Korelasyon Matrisi', fontsize=15, fontweight='bold', pad=15)
    plt.savefig('grafik5_korelasyon.png', dpi=150)
    plt.show()

    # --- GRAFİK 2: REGRESYON (İndentasyon Hatası Giderildi) ---
    X = pd.DataFrame({'BIST100': bist_ret, 'S&P500': sp500_ret, 'USD/TRY': usdtry_ret})
    # Parantez kullanarak güvenli birleştirme (Intersection hatasını çözer)
    common = (X.dropna().index
              .intersection(port_ret.dropna().index))

    X_c, y_c = X.loc[common], port_ret.loc[common]
    X_mat = np.column_stack([np.ones(len(X_c)), X_c.values])
    coefs, _, _, _ = np.linalg.lstsq(X_mat, y_c.values, rcond=None)

    plt.figure(figsize=(10, 5))
    labels = ['Sabit', 'BIST100', 'S&P500', 'USD/TRY']
    plt.barh(labels, coefs, color=STOCK_COLORS[:4])
    plt.title('Regresyon Katsayıları (Portföy Duyarlılığı)', fontsize=14)
    plt.axvline(0, color='white', lw=1)
    plt.savefig('grafik7_regresyon.png', dpi=150)
    plt.show()

    # --- GRAFİK 3: KÜMÜLATİF GETİRİ ---
    plt.figure(figsize=(14, 6))
    cum_port = (1 + port_ret).cumprod()
    cum_bist = (1 + bist_ret).cumprod()
    plt.plot(cum_port, label='Portföy', color=PALETTE['accent1'], lw=2)
    plt.plot(cum_bist, label='BIST100', color=PALETTE['accent2'], ls='--')
    plt.fill_between(cum_port.index, cum_port, cum_bist, alpha=0.1, color='white')
    plt.legend()
    plt.title('Kümülatif Performans Karşılaştırması')
    plt.savefig('grafik8_performans.png', dpi=150)
    plt.show()

# 4. ÇALIŞTIR
run_all_plots()
print("\n✅ Grafikler başarıyla oluşturuldu ve 'Dosyalar' kısmına kaydedildi.")
```

```
!pip install yfinance
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statsmodels.api as sm

# ═══════════════════════════════════════════════════════════════
# 1. VERİ TOPLAMA (DATA COLLECTION)
# ═══════════════════════════════════════════════════════════════
# Metinde belirtilen spesifik semboller
hisseler = ['THYAO.IS', 'GARAN.IS', 'AKBNK.IS', 'ASELS.IS', 'SISE.IS', 'EREGL.IS']
market = 'XU100.IS'        # BIST100 (Yerel Piyasa)
global_mkt = '^GSPC'       # S&P500 (Küresel Piyasa)
fx = 'USDTRY=X'            # USD/TRY (Makroekonomik Koşul)

print("📡 2018-2025 verileri Yahoo Finance üzerinden çekiliyor...")
raw_data = yf.download(hisseler + [market, global_mkt, fx], start='2018-01-01', end='2025-12-31')

# Kapanış fiyatları ve Hacim verilerinin ayrıştırılması
prices = raw_data['Close'].dropna()
volumes = raw_data['Volume'][hisseler].dropna()

# Fiyatları Günlük Getiriye (Daily Returns) Dönüştürme
returns = prices.pct_change().dropna()

# Değişken Grupları
stock_ret = returns[hisseler]
bist_ret = returns[market]
sp500_ret = returns[global_mkt]
usdtry_ret = returns[fx]

# Eşit Ağırlıklı Portföy Getirisi (Equal-Weighted Portfolio)
port_ret = stock_ret.mean(axis=1)

# ═══════════════════════════════════════════════════════════════
# 2. KEŞİFSEL VERİ ANALİZİ (EDA) & GÖRSELLEŞTİRME
# ═══════════════════════════════════════════════════════════════
plt.style.use('dark_background')
fig = plt.figure(figsize=(22, 18), facecolor='#0E1117')
gs = plt.GridSpec(4, 2, figure=fig)

# --- PANEL 1: Kümülatif Performans (Time-Series) ---
ax1 = fig.add_subplot(gs[0, :])
(1 + returns[[market, fx, global_mkt]]).cumprod().plot(ax=ax1, lw=2, alpha=0.8)
(1 + port_ret).cumprod().plot(ax=ax1, color='#00F5D4', lw=4, label='PORTFÖY (Equal-W)')
ax1.set_title("Zaman Serisi: Kümülatif Getiriler (2018-2025)", fontsize=16, fontweight='bold')
ax1.legend(loc='upper left', ncol=4)

# --- PANEL 2: Getiri Dağılımı ve Normallik (Histogram & KDE) ---
ax2 = fig.add_subplot(gs[1, 0])
sns.histplot(port_ret, kde=True, ax=ax2, color='#F15BB5', bins=100)
ax2.set_title(f"Portföy Getiri Dağılımı\n(Skewness: {port_ret.skew():.2f}, Kurtosis: {port_ret.kurt():.2f})", fontsize=14)

# --- PANEL 3: Aykırı Değerler (Boxplot) ---
ax3 = fig.add_subplot(gs[1, 1])
sns.boxplot(data=stock_ret, ax=ax3, palette='viridis')
ax3.set_xticklabels([x.replace('.IS', '') for x in hisseler])
ax3.set_title("Hisse Bazlı Oynaklık ve Aykırı Değer (Outlier) Analizi", fontsize=14)

# --- PANEL 4: Korelasyon Matrisi ---
ax4 = fig.add_subplot(gs[2, 0])
full_corr = pd.concat([stock_ret, bist_ret, sp500_ret, usdtry_ret, port_ret.rename('PORTFOY')], axis=1).corr()
sns.heatmap(full_corr, annot=True, cmap='RdYlGn', fmt='.2f', ax=ax4, center=0)
ax4.set_title("Korelasyon Analizi (Local vs Global vs Macro)", fontsize=14)

# --- PANEL 5: Hacim vs Volatilite ---
ax5 = fig.add_subplot(gs[2, 1])
vol_proxy = port_ret.abs() # Volatilite temsili olarak mutlak getiri
avg_vol = volumes.mean(axis=1)
common = vol_proxy.index.intersection(avg_vol.index)
ax5.scatter(avg_vol.loc[common], vol_proxy.loc[common], alpha=0.3, color='#00BBF9', s=10)
ax5.set_title("İşlem Hacmi - Volatilite İlişkisi", fontsize=14)
ax5.set_xlabel("Ortalama Hacim")
ax5.set_ylabel("|Günlük Getiri|")

# ═══════════════════════════════════════════════════════════════
# 3. HİPOTEZ TESTLERİ & REGRESYON
# ═══════════════════════════════════════════════════════════════
# T-Test: Portföy Getirisi Sıfırdan Farklı mı? (H0: mu = 0)
t_stat_mu, p_val_mu = stats.ttest_1samp(port_ret, 0)

# Paired T-Test: Portföy vs Market (H0: mu_d = 0)
t_stat_pair, p_val_pair = stats.ttest_rel(port_ret, bist_ret)

# Pearson: Hacim vs Volatilite (H0: rho = 0)
corr_v, p_v = stats.pearsonr(avg_vol.loc[common], vol_proxy.loc[common])

# Çoklu Regresyon Analizi (H0: Beta_i = 0)
X = pd.concat([bist_ret, sp500_ret, usdtry_ret], axis=1).dropna()
y = port_ret.loc[X.index]
X = sm.add_constant(X) # Sabit terim (beta_0) ekle
model = sm.OLS(y, X).fit()

# Regresyon Görseli
ax6 = fig.add_subplot(gs[3, :])
model.params[1:].plot(kind='barh', ax=ax6, color=['#00F5D4', '#F15BB5', '#FEE440'])
ax6.set_title(f"Regresyon Katsayıları (R-Squared: {model.rsquared:.2f})", fontsize=14)
ax6.axvline(0, color='white', lw=1)

plt.tight_layout()
plt.show()

# ═══════════════════════════════════════════════════════════════
# 4. SONUÇ RAPORU (SUMMARY)
# ═══════════════════════════════════════════════════════════════
print(f"""
{'='*65}
📊 HİPOTEZ TESTİ VE ANALİZ SONUÇLARI
{'='*65}
1. Ortalama Getiri Testi (H0: mu=0): p-değeri = {p_val_mu:.4f}
   -> {'Sıfırdan anlamlı fark yok (H0 Reddedilemez)' if p_val_mu > 0.05 else 'Sıfırdan anlamlı derecede farklı'}

2. Portföy vs Market Performansı: p-değeri = {p_val_pair:.4f}
   -> {'Marketle benzer performans (H0 Reddedilemez)' if p_val_pair > 0.05 else 'Marketten anlamlı fark var'}

3. Hacim-Volatilite İlişkisi (Pearson): rho = {corr_v:.4f}, p = {p_v:.4e}
   -> {'İstatistiksel olarak anlamlı ilişki var' if p_v < 0.05 else 'İlişki anlamsız'}

4. Çoklu Regresyon Özet (Katsayıların Anlamlılığı):
{model.pvalues.to_string()}

{'='*65}
""")
```

```
!pip install yfinance
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statsmodels.api as sm

# ═══════════════════════════════════════════════════════════════
# 1. VERİ TOPLAMA VE ÖN İŞLEME (DATA COLLECTION)
# ═══════════════════════════════════════════════════════════════
hisseler = ['THYAO.IS', 'GARAN.IS', 'AKBNK.IS', 'ASELS.IS', 'SISE.IS', 'EREGL.IS']
market = 'XU100.IS'
global_mkt = '^GSPC'
fx = 'USDTRY=X'

print("📡 Veriler Yahoo Finance üzerinden çekiliyor (2018-2025)...")
raw_data = yf.download(hisseler + [market, global_mkt, fx], start='2018-01-01', end='2025-12-31')

prices = raw_data['Close'].dropna()
volumes = raw_data['Volume'][hisseler].dropna()
returns = prices.pct_change().dropna()

stock_ret = returns[hisseler]
bist_ret = returns[market]
sp500_ret = returns[global_mkt]
usdtry_ret = returns[fx]
port_ret = stock_ret.mean(axis=1) # Eşit ağırlıklı portföy

# ═══════════════════════════════════════════════════════════════
# 2. İLERİ RİSK ANALİZİ (VaR, DRAWDOWN, BETA)
# ═══════════════════════════════════════════════════════════════
# Value at Risk (VaR %95)
var_95 = np.percentile(port_ret, 5)

# Maximum Drawdown (Maksimum Kayıp)
cum_port = (1 + port_ret).cumprod()
rolling_max = cum_port.cummax()
drawdowns = (cum_port - rolling_max) / rolling_max
max_drawdown = drawdowns.min()

# Rolling Beta (60-Günlük)
window = 60
rolling_beta = port_ret.rolling(window).cov(bist_ret) / bist_ret.rolling(window).var()

# ═══════════════════════════════════════════════════════════════
# 3. GÖRSELLEŞTİRME (PROFESYONEL ANALİZ PANELİ)
# ═══════════════════════════════════════════════════════════════
plt.style.use('dark_background')
fig = plt.figure(figsize=(24, 24), facecolor='#0B0E14')
gs = plt.GridSpec(6, 2, figure=fig, hspace=0.4, wspace=0.2)

# --- A. HİSSE BAZLI PERFORMANS (Tek Tek Çizim) ---
ax1 = fig.add_subplot(gs[0, :])
for s in hisseler:
    ax1.plot((1 + stock_ret[s]).cumprod(), label=s.split('.')[0], lw=1.5, alpha=0.8)
ax1.plot(cum_port, color='white', lw=4, label='PORTFÖY (E-W)', linestyle='--')
ax1.set_title("Hisse Bazlı Kümülatif Getiri (2018-2025)", fontsize=18, fontweight='bold', color='#00F5D4')
ax1.legend(loc='upper left', ncol=7)
ax1.grid(alpha=0.1)

# --- B. HİSSE BAZLI RİSK ANALİZİ (Boxplot) ---
ax2 = fig.add_subplot(gs[1, 0])
sns.boxplot(data=stock_ret, palette='viridis', ax=ax2)
ax2.set_xticklabels([s.split('.')[0] for s in hisseler])
ax2.set_title("Hisse Bazlı Günlük Getiri ve Aykırı Değerler", fontsize=15)

# --- C. PORTFÖYÜN RİSK DAĞILIMI (Histogram & VaR) ---
ax3 = fig.add_subplot(gs[1, 1])
sns.histplot(port_ret, kde=True, color='#F15BB5', bins=100, ax=ax3)
ax3.axvline(var_95, color='red', lw=2, label=f'VaR %95: {var_95:.2%}')
ax3.set_title(f"Portföy Dağılımı (Kurtosis: {port_ret.kurt():.2f})", fontsize=15)
ax3.legend()

# --- D. DINAMİK PİYASA DUYARLILIĞI (Rolling Beta) ---
ax4 = fig.add_subplot(gs[2, :])
ax4.fill_between(rolling_beta.index, rolling_beta, color='#00BBF9', alpha=0.2)
ax4.plot(rolling_beta, color='#00BBF9', lw=2, label='60-Günlük Beta')
ax4.axhline(rolling_beta.mean(), color='white', ls='--', label=f'Ortalama Beta: {rolling_beta.mean():.2f}')
ax4.set_title("Dinamik Piyasa Duyarlılığı (Rolling Beta Analysis)", fontsize=15)
ax4.legend()

# --- E. DRAWDOWN ANALİZİ (Kayıp Dönemleri) ---
ax5 = fig.add_subplot(gs[3, :])
ax5.fill_between(drawdowns.index, drawdowns, 0, color='#FF4D4D', alpha=0.4)
ax5.set_title(f"Portföy Drawdown (Zirveden Düşüş) - MaxDD: %{max_drawdown*100:.2f}", fontsize=15)

# --- F. KORELASYON ISI HARİTASI ---
ax6 = fig.add_subplot(gs[4, 0])
full_corr = pd.concat([stock_ret, bist_ret, sp500_ret, usdtry_ret], axis=1).corr()
sns.heatmap(full_corr, annot=True, cmap='coolwarm', fmt='.2f', ax=ax6, cbar=False)
ax6.set_title("Varlıklar Arası Korelasyon", fontsize=15)

# --- G. HACİM - VOLATİLİTE KÜMELENMESİ ---
ax7 = fig.add_subplot(gs[4, 1])
ax7.scatter(volumes.mean(axis=1), port_ret.abs(), alpha=0.3, color='#FEE440', s=10)
ax7.set_title("İşlem Hacmi vs. Volatilite İlişkisi", fontsize=15)
ax7.set_xlabel("Hacim")
ax7.set_ylabel("|Günlük Getiri|")

# --- H. REGRESYON KATSAYILARI (Faktör Analizi) ---
ax8 = fig.add_subplot(gs[5, :])
X = sm.add_constant(pd.concat([bist_ret, sp500_ret, usdtry_ret], axis=1).dropna())
y = port_ret.loc[X.index]
model = sm.OLS(y, X).fit()
model.params[1:].plot(kind='barh', ax=ax8, color=['#00F5D4', '#F15BB5', '#00BBF9'])
ax8.set_title(f"Regresyon Sonuçları (R-Squared: {model.rsquared:.2f})", fontsize=15)
for i, v in enumerate(model.params[1:]):
    ax8.text(v, i, f' {v:.4f} (p={model.pvalues[i+1]:.3f})', va='center', fontweight='bold')

plt.tight_layout()
plt.show()

# ═══════════════════════════════════════════════════════════════
# 4. HİPOTEZ TESTİ RAPORU
# ═══════════════════════════════════════════════════════════════
print(f"\n{'='*70}\n📋 AKADEMİK ANALİZ VE HİPOTEZ SONUÇLARI\n{'='*70}")
t_stat, p_val = stats.ttest_1samp(port_ret, 0)
print(f"1. Ortalama Getiri Testi (H0: mu=0): p-değeri = {p_val:.4f}")
print(f"2. Risk Analizi: VaR %95 = %{var_95*100:.2f} | Max Drawdown = %{max_drawdown*100:.2f}")
print(f"3. Beta Analizi: Ortalama Beta = {rolling_beta.mean():.2f}")
print(f"4. Regresyon: BIST100 Beta = {model.params[1]:.4f} (p={model.pvalues[1]:.4e})")
print(f"{'='*70}")
```

```
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statsmodels.api as sm
from matplotlib.gridspec import GridSpec

# 1. DATA ACQUISITION & PREPROCESSING
tickers = ['THYAO.IS', 'GARAN.IS', 'AKBNK.IS', 'ASELS.IS', 'SISE.IS', 'EREGL.IS']
market = 'XU100.IS'
macro = 'USDTRY=X'
world = '^GSPC'

print("📡 Connecting to Financial Servers for 2018-2026 Data...")
raw_data = yf.download(tickers + [market, macro, world], start='2018-01-01', end='2026-05-05')['Close']
data = raw_data.ffill().dropna()
returns = data.pct_change().dropna()

# Categorizing returns
stock_ret = returns[tickers]
mkt_ret = returns[market]
fx_ret = returns[macro]
global_ret = returns[world]
port_ret = stock_ret.mean(axis=1) # Equal-Weighted Portfolio

# 2. ADVANCED FINANCIAL CALCULATIONS
# Value at Risk (VaR %95)
var_95 = np.percentile(port_ret, 5)

# Cumulative Returns & Max Drawdown
cum_port = (1 + port_ret).cumprod()
rolling_max = cum_port.cummax()
drawdown = (cum_port - rolling_max) / rolling_max
max_dd = drawdown.min()

# Rolling Beta (60-Day)
window = 60
rolling_beta = stock_ret.mean(axis=1).rolling(window).cov(mkt_ret) / mkt_ret.rolling(window).var()

# 3. SCIENTIFIC VISUALIZATION DASHBOARD
plt.style.use('dark_background')
fig = plt.figure(figsize=(24, 28), facecolor='#0B0E14')
gs = GridSpec(6, 2, figure=fig, hspace=0.35, wspace=0.25)

# --- FIGURE 1: CUMULATIVE PERFORMANCE (Time-Series) ---
ax1 = fig.add_subplot(gs[0, :])
ax1.set_facecolor('#11151C')
for t in tickers:
    ax1.plot((1 + stock_ret[t]).cumprod(), label=t.split('.')[0], alpha=0.6, lw=1)
ax1.plot(cum_port, color='#00F5D4', lw=4, label='EQUITY PORTFOLIO', zorder=5)
ax1.plot((1 + mkt_ret).cumprod(), color='#F15BB5', lw=2, ls='--', label='BIST100 INDEX', alpha=0.8)
ax1.set_title("Figure 1: Cumulative Performance Comparison (2018-2026)", fontsize=18, fontweight='bold', pad=15)
ax1.legend(loc='upper left', ncol=4, frameon=True)
ax1.grid(alpha=0.1)

# --- FIGURE 2: ENHANCED BOXPLOT (Distribution & Outliers) ---
ax2 = fig.add_subplot(gs[1, 0])
ax2.set_facecolor('#11151C')
sns.boxplot(data=stock_ret, palette='viridis', notch=True, ax=ax2,
            flierprops={'marker': 'D', 'markerfacecolor': '#FF3131', 'markeredgecolor': 'white', 'alpha': 0.4})
ax2.set_xticklabels([t.split('.')[0] for t in tickers], rotation=0, fontweight='bold')
ax2.set_title("Figure 2: Daily Return Distribution & Outlier Analysis", fontsize=15, fontweight='bold')
ax2.axhline(0, color='white', ls='--', alpha=0.3)

# --- FIGURE 3: RETURN DENSITY & VaR (Risk Profile) ---
ax3 = fig.add_subplot(gs[1, 1])
ax3.set_facecolor('#11151C')
sns.kdeplot(port_ret, fill=True, color='#00BBF9', ax=ax3, label='Portfolio Return Density')
ax3.axvline(var_95, color='#FF3131', ls='--', lw=2, label=f'95% VaR: {var_95:.2%}')
ax3.set_title(f"Figure 3: Portfolio Risk Profile (Kurtosis: {port_ret.kurt():.2f})", fontsize=15, fontweight='bold')
ax3.legend()

# --- FIGURE 4: DYNAMIC SENSITIVITY (Rolling Beta) ---
ax4 = fig.add_subplot(gs[2, :])
ax4.set_facecolor('#11151C')
ax4.fill_between(rolling_beta.index, rolling_beta, color='#FEE440', alpha=0.15)
ax4.plot(rolling_beta, color='#FEE440', lw=2, label='60-Day Rolling Beta vs BIST100')
ax4.axhline(rolling_beta.mean(), color='white', ls='--', alpha=0.8, label=f'Average Beta: {rolling_beta.mean():.2f}')
ax4.set_title("Figure 4: Dynamic Market Sensitivity (Systematic Risk Evolution)", fontsize=15, fontweight='bold')
ax4.legend()

# --- FIGURE 5: DRAWDOWN ANALYSIS (Peak-to-Trough) ---
ax5 = fig.add_subplot(gs[3, :])
ax5.set_facecolor('#11151C')
ax5.fill_between(drawdown.index, drawdown, 0, color='#FF3131', alpha=0.3)
ax5.plot(drawdown, color='#FF3131', lw=1, alpha=0.7)
ax5.set_title(f"Figure 5: Portfolio Drawdown Analysis (Max Drawdown: {max_dd:.2%})", fontsize=15, fontweight='bold')
ax5.set_ylabel("Loss from Peak")

# --- FIGURE 6: CROSS-ASSET CORRELATION MATRIX ---
ax6 = fig.add_subplot(gs[4, 0])
full_corr = pd.concat([stock_ret, mkt_ret, fx_ret, global_ret], axis=1).corr()
sns.heatmap(full_corr, annot=True, cmap='RdYlGn', fmt='.2f', ax=ax6, cbar=False, center=0)
ax6.set_title("Figure 6: Cross-Asset Correlation Matrix", fontsize=15, fontweight='bold')

# --- FIGURE 7: VOLATILITY CLUSTERING (Hacim vs Risk) ---
ax7 = fig.add_subplot(gs[4, 1])
ax7.set_facecolor('#11151C')
ax7.scatter(volumes.mean(axis=1), port_ret.abs(), alpha=0.4, color='#9b5de5', s=15)
ax7.set_title("Figure 7: Volatility Clustering & Trading Volume Relationship", fontsize=15, fontweight='bold')
ax7.set_xlabel("Average Trading Volume")
ax7.set_ylabel("Absolute Daily Return (Volatility Proxy)")

# --- FIGURE 8: MULTI-FACTOR REGRESSION (Beta Coefficients) ---
ax8 = fig.add_subplot(gs[5, :])
ax8.set_facecolor('#11151C')
X = sm.add_constant(pd.concat([mkt_ret, fx_ret, global_ret], axis=1).dropna())
reg_model = sm.OLS(port_ret.loc[X.index], X).fit()
reg_model.params[1:].plot(kind='barh', ax=ax8, color=['#00F5D4', '#F15BB5', '#9b5de5'], alpha=0.8)
ax8.set_title(f"Figure 8: Factor Sensitivities (R-Squared: {reg_model.rsquared:.2f})", fontsize=15, fontweight='bold')
for i, v in enumerate(reg_model.params[1:]):
    ax8.text(v, i, f' Beta: {v:.4f} (p={reg_model.pvalues[i+1]:.2e})', va='center', fontweight='bold', color='white')

plt.tight_layout()
plt.show()

# FINAL SCIENTIFIC OUTPUTS FOR REPORT
print(f"\n{'='*80}\nACADEMIC SUMMARY STATISTICS\n{'='*80}")
print(f"1. Systematic Risk (Mean Beta): {rolling_beta.mean():.4f}")
print(f"2. Risk at Peril (95% VaR): {var_95:.4%}")
print(f"3. Recovery Metric (Max Drawdown): {max_dd:.4%}")
print(f"4. Portfolio Kurtosis (Tail Risk): {port_ret.kurt():.4f}")
print(f"5. Regression p-values:\n{reg_model.pvalues}")
print(f"{'='*80}")
```
